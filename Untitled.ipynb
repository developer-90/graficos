## Muestra un gr√°fico de barras con el total de productos por categoria

```python
import pandas as pd
import matplotlib.pyplot as plt

def consumir():
    datos = pd.read_csv('products.csv')
    df = datos[['unitsInStock','categoryID']] # convertir datos a dataframe
    valor_productos_categoria = df.groupby('categoryID')['unitsInStock'].mean()
    #valor_productos_categoria.head(5).plot.barh()
    df.plot.barh()
    plt.show()
consumir()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def consumir():
    datos = pd.read_csv('products.csv')
    df = datos[['unitPrice','categoryID']] # convertir datos a dataframe
    df.unitPrice.value_counts().plot.pie()
    plt.show()
consumir()
```
